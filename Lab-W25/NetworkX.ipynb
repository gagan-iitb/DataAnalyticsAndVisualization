#Weighted Graph

```python
WG = nx.Graph()
WG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 0.75), (2, 4, 1.2), (3, 4, 0.375)])

for (u, v, wt) in WG.edges.data('weight'):
	print(f"({u}, {v}, {wt:.3})")

nx.draw(WG,with_labels=True)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a graph
WG = nx.Graph()
WG.add_weighted_edges_from([(1, 2, 0.125), (1, 3, 7.8), (2, 4, 23.4), (3, 4, 3.4),(3,5,2.3)])

# Draw the graph
pos = nx.spring_layout(WG)  # Define layout
nx.draw(WG, pos, with_labels=True)  # Draw nodes and edges
edge_labels = nx.get_edge_attributes(WG, 'weight')  # Get edge weights
nx.draw_networkx_edge_labels(WG, pos, edge_labels=edge_labels)  # Draw edge labels

plt.show()  # Display the graph
```

```python
import networkx as nx

edges = [
    (1, 2, {"weight": 4}),
    (1, 3, {"weight": 2}),
    (2, 3, {"weight": 1}),
    (2, 4, {"weight": 5}),
    (3, 4, {"weight": 8}),
    (3, 5, {"weight": 10}),
    (4, 5, {"weight": 2}),
    (4, 6, {"weight": 8}),
    (5, 6, {"weight": 5}),
]
edge_labels = {
    (1, 2): 4,
    (1, 3): 2,
    (2, 3): 1,
    (2, 4): 5,
    (3, 4): 8,
    (3, 5): 10,
    (4, 5): 2,
    (4, 6): 8,
    (5, 6): 5,
}


G = nx.Graph()
for i in range(1, 7):
    G.add_node(i)
G.add_edges_from(edges)

pos = nx.planar_layout(G)

# This will give us all the shortest paths from node 1 using the weights from the edges.
p1 = nx.shortest_path(G, source=1, weight="weight")

# This will give us the shortest path from node 1 to node 6.
p1to6 = nx.shortest_path(G, source=1, target=6, weight="weight")

# This will give us the length of the shortest path from node 1 to node 6.
length = nx.shortest_path_length(G, source=1, target=6, weight="weight")

print("All shortest paths from 1: ", p1)
print("Shortest path from 1 to 6: ", p1to6)
print("Length of the shortest path: ", length)
```

#Diameter

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an undirected graph
G = nx.Graph()

# Add nodes to the graph
nodes = [1, 2, 3, 4, 5, 6, 7]
G.add_nodes_from(nodes)

# Add edges to the graph
edges = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4), (5, 6), (5, 7)]
G.add_edges_from(edges)

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()

# Perform the operations from the previous code snippet
print(nx.is_connected(G))

components = nx.connected_components(G)
largest_component = max(components, key=len)

subgraph = G.subgraph(largest_component)
diameter = nx.diameter(subgraph)
print("Network diameter of largest component:", diameter)
```

```python
import networkx as nx
import pandas as pd
import io

# Read the CSV file into a pandas DataFrame
# df = pd.read_csv(io.BytesIO(uploaded['graph_data (2).csv']))
df = pd.read_csv('https://raw.githubusercontent.com/gagan-iitb/DataAnalyticsAndVisualization/main/Lab-W25/dataset/graph_data.csv')


# Create an empty graph
G = nx.Graph()

# Iterate over each row in the DataFrame and add edges to the graph
for _, row in df.iterrows():
    source = row['source']
    target = row['target']
    G.add_edge(source, target)

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```
