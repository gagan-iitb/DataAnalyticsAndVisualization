```python
# Import necessary libraries
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)

# Apply PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Visualize the reduced dataset
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target, cmap='viridis')
plt.title('PCA on Iris Dataset (2D Projection)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Iris Species')
plt.show()
```

```python
#Preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
X = iris.data  # Features (sepal length, sepal width, petal length, petal width)

# Apply Standardization (StandardScaler normalizes the data)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Original Data (first 5 samples):")
print(X[:5])  # Original data

print("\nScaled Data (first 5 samples):")
print(X_scaled[:5])  # Scaled data
```

```python
from sklearn.preprocessing import MinMaxScaler

# Apply Min-Max Scaling (scales each feature to [0, 1] range)
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

print("Normalized Data (first 5 samples):")
print(X_normalized[:5])
```

```python
from sklearn.impute import SimpleImputer
import numpy as np

# Create data with missing values
X_missing = np.array([[1, 2, np.nan], [4, np.nan, 6], [7, 8, 9]])

# Impute missing values with the mean of each column
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X_missing)

print("Data with missing values:")
print(X_missing)

print("\nData after imputation (missing values replaced with mean):")
print(X_imputed)
```

```python
from sklearn.feature_selection import VarianceThreshold

# Apply VarianceThreshold to remove features with low variance
selector = VarianceThreshold(threshold=0.1)  # Keep features with variance > 0.1
X_selected = selector.fit_transform(X)

print("Data after removing low-variance features:")
print(X_selected[:5])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Define a preprocessing pipeline
pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),  # Handle missing values
    ('scaler', StandardScaler())  # Standardize features
])

# Apply the pipeline
X_preprocessed = pipeline.fit_transform(X)

print("Preprocessed Data (first 5 samples):")
print(X_preprocessed[:5])
```
