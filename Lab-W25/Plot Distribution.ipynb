Line Plot


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o', color='b')
plt.title('Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

**Bar Plot**

A bar plot is a graphical representation used to compare categorical data. Each category is represented by a bar, and the height (or length) of the bar corresponds to the value or frequency of that category. It is commonly used to showcase comparisons between discrete groups or categories.

```python
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]

plt.bar(categories, values, color='green')
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

**Histogram**

Used for visualizing the distribution of data.

```python
import numpy as np

data = np.random.randn(1000)
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

**Scatter plot**

```python
x = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color='red')
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

**Box plot**

```python
data = [7, 8, 9, 10, 7, 9, 8, 10, 11, 12, 20]

plt.boxplot(data)
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```

**Heatmap**

Used for representing data in a matrix form with varying intensities.

```python
import seaborn as sns
import numpy as np

data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```
